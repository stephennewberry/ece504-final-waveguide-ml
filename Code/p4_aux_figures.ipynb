# Auxiliary Figures
This notebook has the source code for some figures used in the associated report.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configure plot size
plt.rcParams['figure.dpi'] = 90
plt.style.use("seaborn-whitegrid")
plt.rc('font', size=14)
plt.rc('axes', labelsize=24)
```

```python
# Exponential distribution PDF
# From: https://en.wikipedia.org/wiki/Exponential_distribution

x = np.linspace(-1, 5, 1000)
rate_param = [0.5, 1, 2, 3]
y = np.zeros((len(rate_param), len(x)))

for r in range(len(rate_param)):
    for i in range(len(x)):
        if x[i] < 0:
            y[r][i] = 0
        else:
            y[r][i] = rate_param[r] * np.exp(-1*x[i]*rate_param[r])

plt.plot(x, y[0])
plt.plot(x, y[1])
plt.plot(x, y[2])
plt.plot(x, y[3])
plt.legend(["$\lambda=0.5$","$\lambda=1$","$\lambda=2$","$\lambda=3$"])
plt.xlabel("$x$")
plt.ylabel("$P_{x}$")
# Title: Probability density function of the exponentialf
```

```python
y = np.zeros((len(rate_param), len(x)))

for r in range(len(rate_param)):
    for i in range(len(x)):
        if x[i] < 0:
            y[r][i] = 0
        else:
            y[r][i] = 1 - np.exp(-1*x[i]*rate_param[r])

plt.plot(x, y[0])
plt.plot(x, y[1])
plt.plot(x, y[2])
plt.plot(x, y[3])
plt.legend(["$\lambda=0.5$","$\lambda=1$","$\lambda=2$","$\lambda=3$"], loc='lower right')
plt.xlabel("$x$")
plt.ylabel("$P(X \leq x)$")
# Title: Cumulative Distribution function of the exponentialf
```
